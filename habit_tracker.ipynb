```python
# To create an SQL table 
import sqlite3

# To read SQL queries as dataframes
import pandas as pd

# to extract the exact date and time a habit was formed or repeated
from datetime import datetime
```

```python
# creating a databse for the habit tracker
database = 'habits.db'

# connects to the newly established database and executes SQL queries
engine = sqlite3.connect(database)
cursor = engine.cursor()

# creating the columns for the table while informing the user that the table already exists
try:
    cursor.execute("""
               create table habits (
                   id int primary key,
                   habit_name text,
                   date datetime
                   );
               """)
except:
    print('Table already exists')

# saves the table creation 
engine.commit()

print('Database and tables instalized')
```

```python
# function to add values to the table
def add_habits(habit_id, habit_name):

# creates a variable that stores the exact date and time a habit was formed or repeated.
    now = datetime.now().strftime('%Y-%m-%d-%H-%M')

# adds values to the habit tracker (the unique ID, habit, and the datetime it was formed.)
    cursor.execute('insert into habits (id, habit_name, date) values (?, ?, ?);',
                  (habit_id, habit_name, now)
                  )
# saves changes after the code is executed
    engine.commit()
    print(f'{habit_name} add to table successfully on {now}')
```

```python
# adds habit ID, habit name, and date to tracker/table
add_habits(1, 'habit_name')
add_habits(2, 'habit_name')
add_habits(3, 'habit_name')
```
